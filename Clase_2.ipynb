#Cambiar tamaos de la imagen

```
#Expanding
print(img.shape)
res = cv2.resize(img, None, fx=10,fy=10,interpolation=cv2.INTER_CUBIC)
print(res.shape)
plt.imshow(res)
plt.show()
```

```
#SHRINKING
print(img.shape)
res = cv2.resize(img, None, fx=0.1,fy=0.1,interpolation=cv2.INTER_CUBIC)
print(res.shape)
plt.imshow(res)
plt.show()
```

#Remover numeracin de los ejes

```
plt.figure(figsize=(7,7))
plt.imshow(img_BGR)
plt.xticks([])
plt.yticks([])
plt.show()
```

#Cambiar de BGR a RGB

```
#pasar la imagen a rgb

img_RGB = cv2.cvtColor(img_BGR,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(7,7))
plt.imshow(img_RGB)
plt.xticks([])
plt.yticks([])
plt.show()
```

```
#pasar la imagen a GRIS

img_GRAY = cv2.cvtColor(img_BGR,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(7,7))
plt.imshow(img_GRAY, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```
